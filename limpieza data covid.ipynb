```python
import pandas as pd
```

```python
data = pd.read_csv('COVID_DATA-0.csv')
```

```python
data.info()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Drop duplicate rows across all columns
    data = data.drop_duplicates()
    # Drop rows with missing data across all columns
    data = data.dropna(how='all')
    # convert "Edad" column to years
    def convert_to_years(row):
        if row["Unidad de medida de edad"] == 2: # meses
            row["Edad"] = row["Edad"] / 12 
        elif row["Unidad de medida de edad"] == 3: # dias
            row["Edad"] = row["Edad"] / 365
        return row
    data = data.apply(convert_to_years, axis=1)
    # Drop column: 'Unidad de medida de edad'
    data = data.drop(columns=['Unidad de medida de edad'])
    # map column "Pertenencia étnica" (1-Indígena 2-ROM 3-Raizal 4-Palenquero 5-Negro 6-Otro)
    col_map = {
        1: "Indígena",
        2: "ROM",
        3: "Raizal",
        4: "Palenquero",
        5: "Negro",
        6: "Otro",
    }
    data["Pertenencia étnica"] = data["Pertenencia étnica"].map(col_map, na_action="ignore")
    return data

data_clean = clean_data(data.copy())
data_clean.head()
```
